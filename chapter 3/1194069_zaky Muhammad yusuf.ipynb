```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy

plt.rcParams['figure.figsize'] = (8, 8)
```

```python
movies = pd.read_csv('./movies_metadata.csv').dropna()
movies['tagline'] = movies['tagline'].str.lower()
movies.head()

#Dalam latihan ini, Anda telah diberikan kumpulan lebih dari 7000 baris tag film. 
#Tugas Anda adalah membuat bagan representasi kata bow_matrix untuk tagline ini. 
#Untuk latihan ini, kita akan mengabaikan langkah preprocessing teks dan menghasilkan bow_matrix secara langsung.
```

```python
corpus = movies['tagline']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create CountVectorizer object
vectorizer = CountVectorizer()

# Generate matrix of word vectors
bow_matrix = vectorizer.fit_transform(corpus)

# Print the shape of bow_matrix
print(bow_matrix.shape)

#Anda sekarang tahu cara menghasilkan sekantong representasi kata untuk kumpulan dokumen tertentu. 
#Perhatikan bahwa kata vektor yang dibuat memiliki lebih dari 6600 dimensi. 
#Namun, sebagian besar dimensi tersebut memiliki nilai nol karena sebagian besar kata tidak muncul dalam tagline tertentu.
```

```python
nlp = spacy.load('en_core_web_sm')
stopwords = spacy.lang.en.stop_words.STOP_WORDS

#Menganalisis dimensi dan preprocessing
#Dalam latihan ini, Anda telah diberikan lem_corpus yang berisi versi pra-proses dari tagline film dari latihan sebelumnya. 
#Dengan kata lain, tagline telah dikecilkan dan dilemmatisasi, dan stopword telah dihapus.

```

```python
lem_corpus = corpus.apply(lambda row: ' '.join([t.lemma_ for t in nlp(row) 
                                                if t.lemma_ not in stopwords 
                                                and t.lemma_.isalpha()]))

#Tugas Anda adalah membuat bagan representasi kata bow_lem_matrix untuk tagline lemmatisasi ini dan membandingkan
#bentuknya dengan bow_matrix yang diperoleh pada latihan sebelumnya.
```

```python
vectorizer = CountVectorizer()

# Generate of word vectors
bow_lem_matrix = vectorizer.fit_transform(lem_corpus)

# Print the shape of how_lem_matrix
print(bow_lem_matrix.shape)
```

```python
sentences = ['The lion is the king of the jungle',
             'Lions have lifespans of a decade', 
             'The lion is an endangered species']
```

```python
vectorizer = CountVectorizer()

# Generate matrix of word vectors
bow_matrix = vectorizer.fit_transform(sentences)

# Convert bow_matrix into a DataFrame
bow_df = pd.DataFrame(bow_matrix.toarray())

# Map the column names to vocabulary
bow_df.columns = vectorizer.get_feature_names()

# Print bow_df
bow_df

#Memetakan indeks fitur dengan nama fitur
#Dalam video pelajaran, kita telah melihat bahwa CountVectorizer tidak selalu mengindeks kosakata dalam urutan abjad. 
#Dalam latihan ini, kita akan belajar memetakan setiap indeks fitur ke nama fitur yang sesuai dari kosakata.
```
